## A First Look at Neural Networks


```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

The machine learning paradigm is one where you have data, that data is labeled, and you want to figure out the rules that match the data to the labels.  
Consider these two sets of Numpy arrays:

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

When using TensorFlow, we define our layers using `Sequential`.

```python
model = Sequential([Dense(units=1, input_shape=[1])])
```

Inside the `Sequential`, we specify what each layer looks like. Since we have only one line inside our `Sequential`, we have only one layer.  
We then define what the layer look like using the `keras.layers`. `Dense` means a set of fully connected neurons. There are lots of different layer types besides `Dense`.  
Our `Dense` layer has `units=1` specified: So we have just one dense layer with one neuron in our entire neural network.  

In the next line we set our optimizer. The fundamental trick in deep learning is to use a loss function or cost function as a feedback signal to adjust the value of the weights a little, in a direction that will lower the loss score for the current example. It implements whatâ€™s called the Backpropagation algorithm: the central algorithm in deep learning.   

`sgd` stands for *stochastic gradient descent*. A training loop, which, repeated a sufficient number of times, yields weight values that minimize the loss function. Of course, there are other kinds of optimizer. We need just to pick the appropriate one to use for different scenarios.

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

The learning process will then begin with the command `model.fit`:

```python

model.fit(xs, ys, epochs=500);
print(model.predict([10.0]))
```

```python
print(model.predict([10.0]))
```

```python
print("Here is what I learned: {}".format(model.get_weights()))
```
